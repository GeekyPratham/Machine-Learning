Imbalanced Dataset

A dataset with an unequal class distribution

```python
# importing the dependencies
import pandas as pd
import numpy as np
```

```python
#loading the dataset to pandas Dataframe
credit_card_data = pd.read_csv('/content/drive/MyDrive/datasets/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
#distribution of the two classes of dataset
credit_card_data.Class.value_counts()
```

this is highly imbalanced data set as because more than 95% of data belogns to one class

0 -> legit transaction

1-> fraudulent transaction



```python
# seperating the legit and fraudulent transaction
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit)
```

```python
print(fraud)
```

```python
print(legit.shape)
print(fraud.shape)
```

Under-Sampling

taking the data from maximaum data of which group presnt approximatly the lower data

Build a sample dataset containing similar distribution of legit and fraudulent transactions


Number of Fraudulent Transaction - 492

```python
legit_sample = legit.sample(n=492)
```

```python
print(legit_sample)
```

```python
print(legit_sample.shape)
```

#concatenate the two dataframe i.e legit_sample + fraud

```python
new_dataset = pd.concat([legit_sample,fraud], axis = 0)
```
