```python
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
print(y)
print(z)
```

Bar Plot


```python
fig = plt.figure()# this will create an empty plot
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
languages = ['English','Hindi','Bhojpuri','Marathi','Tamil']
people = [100,200,120,30,60]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('No of People')
plt.show()
```

Pi chart

```python
fig1 = plt.figure()
ax1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
languages = ['English','Hindi','Bhojpuri','Marathi','Tamil']
people = [100,200,120,30,60]
ax1.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0.1,0.1,0.8,0.8])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
