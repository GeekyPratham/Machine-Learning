# Arithmetic Operation on two vectors

addition on two vector

```python


vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,3,-2])
sum = vec_1 + vec_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='r')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

 substraction of two vector

```python


vec_1 = np.asarray([0,0,2,3])
vec_2 = np.asarray([0,0,3,-2])
sum = vec_1 - vec_2
print(sum)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='r')

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

 multiplication of vector with a scalar


```python


vec_1 = np.asarray([0,0,2,3])
vec_2 = 4*vec_1

print(vec_2)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,8,12,scale_units='xy',angles='xy',scale=1,color='g')


plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot Product of 2 vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

```python
a1 = np.array([2,3,4])
b1 = np.array([4,4,4])
a1_dot_b1 = np.dot(a1,b1)
```

```python
print(a1_dot_b1)
```

Cross Product of 2 vector

```python
a = np.array([2,3])
b = np.array([4,4])
a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

```python
a1 = np.array([12,3,4])
b1 = np.array([4,4,4])
a1_cross_b1 = np.cross(a1,b1)
print(a1_cross_b1)
```

Projection of a vector on b vector

((a.v)/(|v|)^2)b

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
mag_b = np.sqrt(np.dot(b,b))
magnitude_of_projection = a_dot_b/(mag_b**2)
projection = magnitude_of_projection*b
print(projection)
```
