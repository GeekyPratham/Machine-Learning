### Diabetes Prediction in Female

Importing the dependencies

```python
import pandas as pd # converting dataset into dataframe
import numpy as np # performing arithmetic operation
from sklearn.preprocessing import StandardScaler  # to standardize the data to make all in same range
from sklearn.model_selection import train_test_split # splitting the data into train and test data
from sklearn import svm # importing the support vercotr machine model from sklearn
from sklearn.metrics import accuracy_score # find the accuracy score of model
```

## Data Collection and Analysis

PIMA Diabetes Dataset

```python
# loading the diabetes dataset to the pandas DataFrame

diabetes_dataset = pd.read_csv("/content/drive/MyDrive/datasets/diabetes.csv");
```

```python
# printing first 5 row
diabetes_dataset.head()
```

```python
# printing the last five rows
diabetes_dataset.tail()
```

```python
# getting rows and columns of dataset
diabetes_dataset.shape
```

```python
# printing how many empty shell present
diabetes_dataset.isnull().sum()
```

```python
# printing how many people are diabetic or not
diabetes_dataset.Outcome.value_counts()
```

0 -> non diabetic

1 -> diabetic

```python
# getting statistical measure of data
diabetes_dataset.describe()
```

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
# separating the data and labels

X = diabetes_dataset.drop(columns = 'Outcome',axis=1)
Y = diabetes_dataset.Outcome
```

```python
X.head()
```

```python
Y.head()
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X) # this will give numpy array
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset.Outcome
```

```python
print(X)
print(type(X))
```

```python
print(Y)
print(type(Y))
```

Train Test Split

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

Training the Model

```python
classifier = svm.SVC(kernel='linear')# we are using linear model
```

```python
# training the support vector Machine Classifier
classifier.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score on Training data

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy score on the training data")
print(training_data_accuracy)
```

Accuracy Score on Test data

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score on the Test data")
print(test_data_accuracy)
```
