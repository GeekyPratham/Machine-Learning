Manipulating a DataFrame

```python
# adding a column to a dataframe

# in diabetes dataframe there is no marital status for lady so i take 3 option (Married,unMarried,Divorced)
# adding the new column to this dataframe with random Marital Status

# Define the options
status_options = ['Married', 'UnMarried', 'Divorced']

# Assign random statuses to each row

diabetes_df['Marital_Status'] = np.random.choice(status_options, size=len(diabetes_df))
```
