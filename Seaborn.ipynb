Seaborn:

 Data Visualization Library

Importing the libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
print(tips.head())
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

Count Plot


```python
sns.countplot(x='class',data = titanic)


```

```python
sns.countplot(x='survived',data = titanic)
```

Bar Chart


```python
sns.barplot(x='sex',y = 'survived',hue='class',data = titanic)
```

Distribution Plot

```python
sns.displot(titanic['age'])
```

```python
sns.distplot(titanic['age'])
```

correlation Matrix

  Positive Correlation
  
  Negative Correlation

```python
# consturcting Heat Map for correlation

correlation = titanic.corr(numeric_only=True)
print(correlation)

plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```
