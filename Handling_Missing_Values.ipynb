Methods to Handle Missing Values:

 1. Imputation -> replacing the missing (NaN) with the statistical value(mean,median,mode)
 2. Dropping -> it means dropping all the rows which has missing values and this is usually doen when we have a large no of dataset

Importing the libraries

Central Tendencies
1. Mean
2. Median
3. Mode

```python
# analysing the distribution of data in salary
fig , ax = plt.subplots(figsize=(8,8))
sns.distplot(dataset.salary)
```

```python
# analysing the distribution of data in salary
fig , ax = plt.subplots(figsize=(8,8))
sns.histplot(dataset.salary)
```

in case of skew graph/plot we do not replace the missing value with mean value because in skew graph thre is may density in particular region and few are present with high value this can affect the overall  mean. so in this canse we can go replace with mode or median

```python
dataset['salary'].fillna(dataset['salary'].median(),inplace=True)
display(dataset.head())
```
