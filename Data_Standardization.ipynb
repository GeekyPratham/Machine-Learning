Data Standarddization

The process of standardizing  the data to a common format and common range

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset
dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(dataset)
```

```python
# loading the data to a pandas dataframe

df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
df.head()
```

# from above details of dataset(df) we can understand that some of the columns value ranges form (10-20) and some ranges form 10-200 and some ranges form 1000-2000 and some are less than 0  so we need to standardize data for better processing and analysing

```python
df.shape
```

```python
x = df
y = dataset.target
```

```python
print (x)
print(y)
```

Splitting the data into training data and test data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

Standardize the data

```python
print(dataset.data.std())
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_standardize = scaler.transform(x_train)
```

```python
print(x_train_standardize)
```

```python
data = pd.DataFrame(x_train_standardize,columns = dataset.feature_names)
```

```python
print(data)
```

```python
data.head()
```

```python
x_test_standardize = scaler.transform(x_test)
```

```python
print(x_test_standardize)
```

```python
data2 = pd.DataFrame(x_test_standardize,columns = dataset.feature_names)
```

```python
print(data2)
```
